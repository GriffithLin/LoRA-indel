```python
def get_dataList(fasta_file, out_file):
    data = pd.read_csv(fasta_file, header=None)
    data_new = data[1::2]
    data_new = data_new.rename(columns={0: 'protein_seq'})
    data_new = data_new.reset_index(drop=True)
    data_new["label"] = data[::2][0].apply(lambda x:int(x[-1])).reset_index(drop=True)
    data_new["label"].value_counts()
#     print(data_new)
    data_new.to_csv(out_file)
# 
```

```python
def to_vcf(data_list, out_file):
    data_list["ID"] = '.'
    data_list["QUAL"] = '.'
    data_list["FILTER"] = '.'
    data_list.to_csv(out_file,index=False,mode="w",na_rep=".",sep="\t",columns=["#CHROM", "POS", "ID", "REF", "ALT", "QUAL", "FILTER" ,"INFO"])
```

```python
get_dataList("test.fasta", "test_protein.csv")
get_dataList("train.fasta", "train_protein.csv")
get_dataList("DDD.fasta", "DDD_protein.csv")
```

```python
get_dataList("target_test.fasta", "target_test_protein.csv")
get_dataList("target_data.fasta", "target_data_protein.csv")
```
